```python
data = pd.read_csv("obverse_features.csv")
features_ = data.iloc[:, 1:].values
```

```python
features_.shape
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Tune eps and min_samples for your dataset
labels = dbscan.fit_predict(features_scaled)
```

```python

# Visualize using PCA
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)
```

```python

plt.figure(figsize=(10, 7))
scatter = plt.scatter(features_pca[:, 0], features_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('DBSCAN Clustering Visualization')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import MeanShift
```

```python
# Apply Mean Shift
mean_shift = MeanShift()
labels = mean_shift.fit_predict(features_scaled)
```

```python
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(features_pca[:, 0], features_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('Mean Shift Clustering Visualization')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(distance_threshold=0, n_clusters=None, compute_full_tree=True)
labels = agg_clustering.fit_predict(features_scaled)
```

```python

# Visualize using PCA
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(features_pca[:, 0], features_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('Agglomerative Clustering Visualization')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid()
plt.show()
```

```python
# Visualize Dendrogram
linkage_matrix = linkage(features_scaled, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import matplotlib.pyplot as plt

# Sample a subset of your data
sample_size = 100  # Choose a reasonable subset size
np.random.seed(42)  # For reproducibility
sample_indices = np.random.choice(features_scaled.shape[0], size=sample_size, replace=False)
sample_data = features_scaled[sample_indices]

# Compute the linkage matrix and plot the dendrogram
linkage_matrix = linkage(sample_data, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, labels=sample_indices)
plt.title('Dendrogram (Sampled Data)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
